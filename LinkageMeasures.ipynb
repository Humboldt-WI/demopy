# Linkage Measures
In this demo, we illustrate three popular linkage measures, which are routinely used in the scope of hierarchical clustering. Generating synthetic data with two well-separable clusters using the function `sklearn.datasets.make_blobs()`, we perform agglomerative clustering using each of the above linkage measures and plot the resulting dendrograms.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Generate synthetic data
X, y = make_blobs(n_samples=50, centers=2, n_features=2, random_state=888, cluster_std=2)

# Plot original data
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=y, marker="x")
plt.title("Original Data with Two Clusters")
plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.show()

# Different linkage measures
linkage_measures = ['ward', 'single', 'average', 'complete']

# Create dendrogram for each linkage measure
for measure in linkage_measures:
    # Perform hierarchical clustering
    linked = linkage(X, method=measure)
    
    # Plot dendrogram
    plt.figure(figsize=(10, 5))
    dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
    plt.title(f'Dendrogram using {measure.capitalize()} Linkage')
    plt.xlabel("Sample index")
    plt.ylabel("Distance")
    plt.show()
```
