# Linkage Measures
In this demo, we illustrate three popular linkage measures, which are routinely used in the scope of hierarchical clustering. Generating synthetic data with two well-separable clusters using the function `sklearn.datasets.make_blobs()`, we perform agglomerative clustering using each of the above linkage measures and plot the resulting dendrograms.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Create synthetic data
X, y = make_blobs(n_samples=50, centers=2, random_state=888, cluster_std=2)

# Plot original data
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')
plt.title('Synthetic')
plt.ylabel('$X_2$')
plt.xlabel('$X_1$')
plt.show()

# Define linkage measures
linkage_measures = ['ward', 'single', 'average']

# Run hierarchical clustering and plot dendrograms
for measure in linkage_measures:
    # Run hierarchical clustering
    model = AgglomerativeClustering(linkage=measure)
    model = model.fit(X)
    
    # Create linkage matrix for dendrogram
    Z = linkage(X, method=measure)
    
    # Plot dendrogram
    plt.figure(figsize=(10, 4))
    dendrogram(Z)
    plt.title(f'Dendrogram ({measure.capitalize()} Linkage)')
    plt.show()
```
