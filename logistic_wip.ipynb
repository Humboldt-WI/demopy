# The Logistic Function

The famous logistic function is mathematically defined as 

$$ g(z) = \frac{1}{1+\exp \left(-z \right)} $$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic(z):
    """ Shamelessly simply implementation 
        of the logistic function.

        Input: z (number array of input values)

        Output: value of the logistic function
                (array of same length as z)
    """
    return 1 / (1 + np.exp(-z))
```

## Helper function for plotting

```python
def plot_logit_function(input_value, scaling=None, shifting=None, flip=False):
    """ Uses the implementation of the logistic function to create a plot of the function """

    if flip==True:
        input_value = -1*input_value
        
    logit_output = logistic(-input_value)

    show_legend = False
    
    plt.plot(input_value, logit_output, marker="+", markeredgecolor="black",
             label="Logistic function ($w={:.1f}$)".format(1.0))
    plt.grid(True)
    plt.xlabel("input values")
    plt.ylabel("Value of logistic function")

    if scaling is not None:
        show_legend = True
        for s in scaling:
            plt.plot(input_value, logistic(s*logit_input), linestyle="dashed",
                     label="Logistic function ($w={:.1f}$)".format(s))

    if shifting is not None:
        show_legend = True
        for s in shifting:
            plt.plot(input_value, logistic(s+logit_input), linestyle="dotted",
                     label="Logistic function ($b={:.1f}$)".format(s))
    
    if show_legend==True:
        plt.legend(loc="best")
    plt.show()
```

```python
logit_input = np.linspace(start=-5, stop=5, num=25)
plot_logit_function(logit_input, flip=True)
```

## Scaling
We can scale the logistic function by multiplying it with some factor. Effectively, we now compute

$$ g(z) = \frac{1}{1+\exp \left(-w \cdot z \right)} $$, where $w$ is the scaling factor.

```python
plot_logit_function(logit_input, scaling=[0.2, 0.5, 2, 4])
```

## Shifting the function
We can also shift the function by adding a constant. Now we compute

$$ g(z) = \frac{1}{1+\exp \left(b -z \right)} $$, where $b$ denotes the constant. Note how this does not alter the shape of the function. We simply push it to the left or to the right.

```python
plot_logit_function(logit_input, shifting=[-2, -1, 1, 2])
```
