# Squared error and quad-quad loss
Simply example of the shape of the squared error (SE) loss function compared to the so-called quad-quad cost function (QQC) (see, e.g., [Dress et al. 2017](https://arxiv.org/abs/1707.02736))

```python
# Import relevant libraries
import numpy  as np
import matplotlib.pyplot as plt

# Generate values of the residual 
x = np.linspace(start=-4, stop=4, num=100)

# Compute squared error loss
se = np.array(x**2)

a = 0.8  # weight of negative residuals in QQC
b = 0.3  # weight of positive residuals in QQC
# Compute quad quad loss
qqc = np.concatenate((a*se[x<=0], b*se[x>0]), axis=0) 

# Plotting
plt.plot(x, se, "-", c="black", label="SE")
plt.plot(x, qqc, "--", c="red",  label="QQC")

plt.xlabel("$\epsilon$")
plt.ylabel("Forecast Error (Cost)")
plt.legend(loc='best')
plt.show()
```
