# Squared error and quad-quad loss
Simply example of the shape of the squared error loss function compared to the so-called quad-quad loss function (see, e.g., [Dress et al. 2017](https://arxiv.org/abs/1707.02736))

```python
# Import relevant libraries
import numpy  as np
import matplotlib.pyplot as plt

# Generate values of the residual 
x = np.linspace(start=-4, stop=4, num=100)

# Compute squared error loss
y_se = np.array(x**2)

a = 0.25  # weight of positive residuals in quad-quad loss
# Compute quad quad loss
y_qq = np.concatenate((y_se[x<0], a*y_se[x>=0]), axis=0) 

# Plotting
plt.plot(x, y_se, "-", c="black", label="Squared error")
plt.plot(x, y_qq, "--", c="red",  label="QQ loss")

plt.xlabel("$\epsilon$")
plt.ylabel("Forecast Error")
plt.legend(loc='best')
plt.show()
```
