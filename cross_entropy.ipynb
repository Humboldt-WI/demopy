# Cross Entropy Loss Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cross_entropy_loss(y, y_hat, ets=1e-15):
    """ Binary cross-entropy loss function

    """
    return -np.sum(y * np.log(y_hat+ets))
```

```python
def cross_entropy_loss(y, y_hat):
    """ Binary cross-entropy loss function

    """
    return -np.sum(y * np.log(y_hat))
```

```python
# Generate some sample data
y = np.array([1, 0])
y_hat = np.linspace(0.0, 1, 20)

# Calculate cross entropy loss for each y_hat
loss = [cross_entropy_loss(y, np.array([yh, 1-yh])) for yh in y_hat]

# Plot the loss function
plt.figure(figsize=(8,4))
plt.plot(y_hat, loss, marker="+")
plt.xlabel('Predicted Probabilities ($\hat{Y}$)')
plt.ylabel('Cross Entropy Loss ($Y=1$)')
#plt.title('Cross Entropy Loss Function')
plt.xticks([p for p in y_hat], ["[{:.1f},{:.1f}]".format(p,1-p) for p in y_hat], rotation=70)
plt.grid(False)
plt.show()

    
```
