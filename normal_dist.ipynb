```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Mean and standard deviation
mu = 0
sigma = 1

# Create a range
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Create the normal distribution for the range
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10,5))

# Plot the normal distribution
plt.plot(x, y)

# Highlight the mean
plt.axvline(mu, color='r', linestyle='dashed', linewidth=2)

# Define colors for different standard deviations
colors = ['blue', 'green', 'purple']

# Highlight areas within one, two and three standard deviations
for i in range(1, 4):
    plt.fill_between(x, y, where=((mu-i*sigma <= x) & (x <= mu+i*sigma)), color=colors[i-1], alpha=0.2)

# Set x-ticks
xticks = [mu - i*sigma for i in range(4, -1, -1)] + [mu + i*sigma for i in range(1, 5)]
plt.xticks(xticks, [f'$\mu-{i}\sigma$' if i != 0 else '$\mu$' for i in range(4, -1, -1)] + [f'$\mu+{i}\sigma$' for i in range(1, 5)])

plt.title('Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Mean and standard deviation for standard normal distribution
mu = 0
sigma = 1

# Create a range
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Create the normal distribution for the range
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10,5))

# Plot the normal distribution
plt.plot(x, y)

# Highlight the median
plt.axvline(mu, color='r', linestyle='dashed', linewidth=2)

# Calculate the first and third quartiles
q1 = norm.ppf(0.25, mu, sigma)
q3 = norm.ppf(0.75, mu, sigma)

# Add vertical dotted lines for the first and third quartiles
plt.axvline(q1, color='g', linestyle='dotted', linewidth=2)
plt.axvline(q3, color='b', linestyle='dotted', linewidth=2)

# Fill the area of the plot corresponding to the interquartile range
plt.fill_between(x, y, where=((q1 <= x) & (x <= q3)), color='yellow', alpha=0.3)

plt.title('Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Standard normal distribution parameters
mu = 0
sigma = 1

# Generate points on the x axis
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the standard normal distribution
plt.plot(x, y, label='Standard Normal Distribution', c='black')

# Fill the area between z = -3 and z = 3
plt.fill_between(x, y, where=(x >= -3) & (x <= 3), color='blue', alpha=0.1, label='$P(-3  Z  3)$')
plt.fill_between(x, y, where=(x < -3) | (x > 3), color='orange', alpha=1, label='$P(|Z|>3)$')

# Annotate mu, mu  sigma, mu  2sigma, etc.
plt.axvline(x=mu, color='grey', linestyle='--')
plt.axvline(x=mu + sigma, color='grey', linestyle='--')
plt.axvline(x=mu - sigma, color='grey', linestyle='--')
plt.axvline(x=mu + 2*sigma, color='grey', linestyle='--')
plt.axvline(x=mu - 2*sigma, color='grey', linestyle='--')
plt.axvline(x=mu + 3*sigma, color='red', linestyle='--')
plt.axvline(x=mu - 3*sigma, color='red', linestyle='--')

plt.text(mu, 0.1, '$\mu$', horizontalalignment='center')
plt.text(mu + sigma, 0.1, '$\mu + \sigma$', horizontalalignment='center')
plt.text(mu - sigma, 0.1, '$\mu - \sigma$', horizontalalignment='center')
plt.text(mu + 2*sigma, 0.1, '$\mu + 2\sigma$', horizontalalignment='center')
plt.text(mu - 2*sigma, 0.1, '$\mu - 2\sigma$', horizontalalignment='center')
plt.text(mu + 3*sigma, 0.1, '$\mu + 3\sigma$', horizontalalignment='center', color='red')
plt.text(mu - 3*sigma, 0.1, '$\mu - 3\sigma$', horizontalalignment='center', color='red')

# Calculate the probability of Z being between -3 and +3
p_in_range = stats.norm.cdf(3) - stats.norm.cdf(-3)

# Annotate probability regions
plt.annotate(f'$P(-3 < Z  3)  {p_in_range:.3f}$', xy=(0, 0.25), xytext=(-4, 0.3), color='blue',
             arrowprops=dict(facecolor='blue', shrink=0.05))

plt.annotate(f'$P(|Z|>3)  {1-p_in_range:.3f}$', xy=(3.1, 0), xytext=(1, 0.2), color='orange',
             arrowprops=dict(facecolor='orange', shrink=0.05))


# Set labels and title
plt.xlabel('Z-score')
plt.ylabel('Density')
#plt.title('Density of Standard Normal Distribution')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
